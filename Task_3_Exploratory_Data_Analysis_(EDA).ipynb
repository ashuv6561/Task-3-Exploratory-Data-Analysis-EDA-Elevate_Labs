```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Load dataset
iris_data = load_iris()
iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris['species'] = iris_data.target

# Histograms
iris.hist(figsize=(10,8), bins=20)
plt.suptitle("Distribution of Iris Features")
plt.show()
```

Insight:
- Petal length and petal width show bimodal distributions  strong species separation.
- Sepal features overlap more, less discriminative.


2. Categorical Features (Count Plot)

```python
sns.countplot(x=iris['species'])
plt.title("Species Count")
plt.show()
```

Insight:
- Dataset is balanced (50 samples per species).


3. Outlier Detection (Box Plots)

```python
for col in iris_data.feature_names:
    sns.boxplot(x="species", y=col, data=iris)
    plt.title(f"Boxplot of {col} by Species")
    plt.show()
```

Insight:
- Sepal width shows mild outliers.
- Petal features clearly separate setosa from versicolor and virginica.


4. Correlation Heatmap

```python
sns.heatmap(iris.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

Insight:
- Petal length  petal width: strong positive correlation.
- Sepal length  sepal width: weak correlation.


5. Key Insights per Visualization
- Histograms  Petal features show clear species separation.
- Count plot  Balanced dataset, no class imbalance.
- Box plots  Outliers mainly in sepal width.
- Heatmap  Petal features are highly correlated and predictive.


6. Important Features for Prediction
- Petal length and petal width are the most discriminative features.
- Sepal features overlap across species, less useful for classification.



7. Summary of Findings
- Iris dataset has 150 samples, evenly distributed across 3 species.
- Petal measurements are the strongest predictors of species.
- Sepal measurements overlap, weaker predictors.
- Mild outliers exist in sepal width.
- Strong correlation between petal length and petal width.
